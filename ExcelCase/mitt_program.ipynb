```python
import pandas as pd
```

```python
sheet1 = pd.read_excel('movies.xls', sheet_name='1900s')
sheet2 = pd.read_excel('movies.xls', sheet_name='2000s')

movies = pd.concat([sheet1, sheet2])
```

```python
movies.head(10)
```

```python
movies.shape
```

```python
movies.describe()
```

```python
directors = movies.groupby(["Director"])['IMDB Score'].mean().sort_values(ascending=False)
```

```python
directors
```

```python
movies["Net Earnings"] = movies["Gross Earnings"] - movies["Budget"]
```

```python
output = movies[["Title", "Budget", "Gross Earnings", "Net Earnings"]].sort_values(by="Net Earnings", ascending=False).head(20)
```

```python
output
```

```python
output.to_excel("output.xls")
```
